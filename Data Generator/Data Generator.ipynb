```python
from faker import Faker
import psycopg2
import random

#Mechanics entity data generation

# Connect to the PostgreSQL database
conn = psycopg2.connect(
    host="localhost",
    database="carsdb",
    user="carsuser",
    password="1234"
)

# Create a cursor object to execute SQL queries
cur = conn.cursor()

# Initialize Faker with a seed to generate consistent data
fake = Faker()
Faker.seed(1234)

# Define a list of car brands to choose from
car_brands = ['Toyota', 'Honda', 'Ford', 'Chevrolet', 'Nissan', 'Jeep']

# Define the number of records to generate
num_records = 1000000

sql = "DELETE FROM mechanics;"
cur.execute(sql)

# Loop to generate and insert records
for i in range(num_records):
    # Generate fake data for each record
    mechanicid = i + 1
    name = fake.name()
    address = fake.address()
    cnp = fake.random_int(min=100000, max=999999)
    experience = random.randint(1, 30)
    favourite_car_brand = random.choice(car_brands)

    # Construct the SQL query to insert the record into the Mechanics table
    sql = f"INSERT INTO Mechanics (mechanicid, name, cnp, address, experience, favourite_car_brand) VALUES ('{mechanicid}', '{name}', '{cnp}', '{address}', {experience}, '{favourite_car_brand}');"

    # Execute the query
    cur.execute(sql)

# Commit the changes to the database
conn.commit()

# Close the cursor and database connection
cur.close()
conn.close()
```

```python
from faker import Faker
import psycopg2
import random

#Cars entity data generation

# Connect to the PostgreSQL database
conn = psycopg2.connect(
    host="localhost",
    database="carsdb",
    user="carsuser",
    password="1234"
)

# Create a cursor object to execute SQL queries
cur = conn.cursor()

# Initialize Faker with a seed to generate consistent data
fake = Faker()
Faker.seed(1234)

# Define a list of car brands to choose from
car_Manufacturer = ['Toyota', 'Honda', 'Ford', 'Chevrolet', 'Nissan', 'Jeep']
car_Model = ['Model1', 'Model2', 'Model2', 'Model3', 'Model4', 'Model5', 'Model6', 'Model7']

# Define the number of records to generate
num_records = 1000000

sql = "DELETE FROM cars;"
cur.execute(sql)

# Loop to generate and insert records
for i in range(num_records):
    # Generate fake data for each record
    carID = i + 1
    dealershipID = i + 1
    CarManufacturer = random.choice(car_Manufacturer)
    CarModel = random.choice(car_Model)
    CarRetailPrice = round(random.uniform(10000.00, 100000.00), 2)
    CarTopSpeed = random.randint(140, 400)
    CarWeight = round(random.uniform(1000.00, 4000.00), 2)

    # Construct the SQL query to insert the record into the Mechanics table
    sql = f"INSERT INTO cars (carid, car_manufacturer, car_model, car_retail_price, car_top_speed, car_weight, dealershipid) VALUES ('{carID}', '{CarManufacturer}', '{CarModel}', '{CarRetailPrice}', {CarTopSpeed}, '{CarWeight}', '{dealershipID}');"

    # Execute the query
    cur.execute(sql)

# Commit the changes to the database
conn.commit()

# Close the cursor and database connection
cur.close()
conn.close()
```
